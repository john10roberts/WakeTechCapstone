```python
import pandas as pd
import numpy as np
from sklearn import metrics
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('C:/manju/waketech/projects/wakeprop/2017/wakesubset1.csv',low_memory=False)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['basementpct2']=(df['basementpct']*0.01)
df['basementpct2'].describe()
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df['zipcode2']=df['zipcode'].astype(str)
df['testcat']=df['testdat'].astype(str)
df['UNITS2']=df['UNITS'].astype(str)
df.drop(['zipcode','testdat','UNITS'],axis=1,inplace=True)
df.describe()
```

```python
df.isnull().sum()
```

```python
df['UTILITIES'].value_counts()
```

```python
df['UTILITIES'] = df['UTILITIES'].fillna(df['UTILITIES'].mode()[0])
df['UTILITIES'].value_counts()
```

```python
df2=df[['STORY_HEIGHT','basementpct2','HEAT','EXTERIOR_WALL','AIR','BATH','BUILTIN1','UNITS2','UTILITIES',
           'GRADE','PHYSICAL_CITY','zipcode2','lprice','lsqft','larea','lage','testcat','REAL_ESTATE_ID','ACCR_COND_PERCENT_ASSD']]
df2.shape
```

```python
df2.STORY_HEIGHT.value_counts()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
```

```python
df3.shape
```

```python
dfy = df3['lprice']
dfx=df3.drop('lprice',1)
dfx.dtypes
```

```python
dfx=pd.get_dummies(dfx)
scaler=StandardScaler()
continuousvars=['larea','lsqft','lage','ACCR_COND_PERCENT_ASSD']
scaler.fit(dfx[continuousvars])
dfx[continuousvars] = scaler.transform(dfx[continuousvars])
#print('With Dummies',list(dfx))
dfx.shape
```

```python
tst=(df3['testcat']=='1')
trn=(df3['testcat']=='0')
xtr=dfx.loc[trn]
xts=dfx.loc[tst]
ytrn=dfy.loc[trn]
ytst=dfy.loc[tst]
xtrn=xtr.drop('REAL_ESTATE_ID',axis=1)
xtst=xts.drop('REAL_ESTATE_ID',axis=1)
xtrn.shape
```

```python
def TestModel(algo):
    algo.fit(xtrn, ytrn)
    return algo.__class__.__name__, metrics.mean_squared_error(algo.predict(xtst), ytst), metrics.r2_score(ytst,algo.predict(xtst)), algo.coef_
an, am, ar, ac = TestModel(LinearRegression())
bn, bm, br, bc = TestModel(LassoCV(alphas=[0.005, 0.001, 0.01, 0.1, 1, 10], normalize=False, random_state=1))
cn, cm, cr, cc = TestModel(RidgeCV(alphas=10.**np.arange(-2, 3), normalize=False, scoring='neg_mean_squared_error'))

coef = pd.DataFrame({'Feature':list(xtrn),an:ac,bn:bc,cn:cc})
mse = pd.DataFrame({'Method':[an,bn,cn],'MSE':[am,bm,cm]})
rsq = pd.DataFrame({'Method':[an,bn,cn],'RSq':[ar,br,cr]})
print(coef)
print(mse)
print(rsq)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=500, oob_score=True, random_state=0)
rf.fit(xtrn, ytrn)
```

```python
from sklearn.metrics import r2_score
predicted_train = rf.predict(xtrn)
predicted_test = rf.predict(xtst)
test_score = r2_score(ytst, predicted_test)
print(test_score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import spearmanr, pearsonr
gbr=GradientBoostingRegressor(n_estimators=400,max_depth=5,min_samples_split=2,learning_rate=0.1,loss='ls')
gbr.fit(xtrn,ytrn)
predicted_train = gbr.predict(xtrn)
predicted_test = gbr.predict(xtst)
test_score = r2_score(ytst, predicted_test)
print("Test data R-2 score", test_score)
```

```python
feature_names=xtrn.columns
featurelist=pd.DataFrame(feature_names,gbr.feature_importances_)
print (featurelist)
```
